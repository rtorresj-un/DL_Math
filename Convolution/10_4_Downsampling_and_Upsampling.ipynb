```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
```

```python
def subsample(x_in):
  x_out = np.zeros(( int(np.ceil(x_in.shape[0]/2)), int(np.ceil(x_in.shape[1]/2)) ))
  # TO DO -- write the subsampling routine
  # Replace this line
  x_out = x_in[::2, ::2]


  return x_out
```

```python
# Now let's try max-pooling
def maxpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  # TO DO -- write the maxpool routine
  # Replace this line
  for i in range(int(np.floor(x_in.shape[0]/2))):
      for j in range(int(np.floor(x_in.shape[1]/2))):
          x_out[i, j] = np.max(x_in[2 * i:2 * i + 2, 2 * j:2 * j + 2])

  return x_out
```

```python
# Finally, let's try mean pooling
def meanpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  # TO DO -- write the meanpool routine
  # Replace this line
  for i in range(int(np.floor(x_in.shape[0]/2))):
      for j in range(int(np.floor(x_in.shape[1]/2))):
          block = x_in[i*2:i*2 + 2, j*2:j*2 + 2]
          x_out[i, j] = np.mean(block)
  
  return x_out
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[6, 8], [8,4]])
print(orig_2_2)
```

```python
# Let's first use the duplication method
def duplicate(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TO DO -- write the duplication routine
  # Replace this line
  for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            x_out[2 * i, 2 * j] = x_in[i, j]
            x_out[2 * i, 2 * j + 1] = x_in[i, j]
            x_out[2 * i + 1, 2 * j] = x_in[i, j]
            x_out[2 * i + 1, 2 * j + 1] = x_in[i, j]

  return x_out
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in, x_high_res):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TO DO -- write the subsampling routine
  # Replace this line
  for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            # Find the position of the maximum value in the original high-res image
            max_value = x_in[i, j]
            # Place the maximum value into the appropriate 2x2 block in the output
            # Find the corresponding position in the high-res image
            for di in range(2):
                for dj in range(2):
                    # We use the position of the maximum value in the high-res image
                    # The exact position in the high-res image is deduced from the pooling
                    if x_high_res[2 * i + di, 2 * j + dj] == max_value:
                        x_out[2 * i + di, 2 * j + dj] = max_value

  return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  x_in_pad = np.zeros((x_in.shape[0]+1, x_in.shape[1]+1))
  x_in_pad[0:x_in.shape[0],0:x_in.shape[1]] = x_in
  # TO DO -- write the duplication routine
  # Replace this line
  # Iterate over the output array
  for i in range(x_out.shape[0]):
        for j in range(x_out.shape[1]):
            # Find the corresponding position in the padded input
            x = i / 2.0
            y = j / 2.0
            
            # Use np.ceil to handle fractional values
            x0 = int(np.floor(x))
            x1 = int(np.ceil(x))
            y0 = int(np.floor(y))
            y1 = int(np.ceil(y))

            # Ensure indices are within bounds
            x1 = min(x1, x_in_pad.shape[0] - 1)
            y1 = min(y1, x_in_pad.shape[1] - 1)

            # Calculate the weights for the four surrounding pixels
            wx = x - x0
            wy = y - y0
            w00 = (1 - wx) * (1 - wy)
            w01 = (1 - wx) * wy
            w10 = wx * (1 - wy)
            w11 = wx * wy

            # Compute the interpolated value
            x_out[i, j] = (w00 * x_in_pad[x0, y0] +
                           w01 * x_in_pad[x0, y1] +
                           w10 * x_in_pad[x1, y0] +
                           w11 * x_in_pad[x1, y1])

  return x_out
```
